```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 예제 1

```python
X = np.array([[1, 1], [1, 2], [1, 3], [1, 4]])
Y = np.array([1.2, 1.8, 3.3, 4.1])
# np.linalg.inv(): 역 행렬(^-1)
# X.T: X의 전치 행렬
# np.dot(): 행렬 곱셈(dot product)
# X.dot(A): X와 A의 행렬 곱셈
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)

# plot을 통한 시각화
plt.scatter(X[:, 1], Y)
plt.plot(X[:, 1], X.dot(theta))
plt.show()
```

```python
cost = (Y - X.dot(theta)) ** 2
print(cost)
print(cost.sum() / len(X))
```

#### 예제 2

```python
X = np.array([[1, 1], [1, 1.2], [1, 1.3], [1, 1.4], [1, 2], [1, 3], [1, 4], [1, 5], [1, 6], [1, 7]])
Y = np.array([1, 1, 1, 1, 4, 5, 5.5, 5.8, 6, 6.1])
# np.linalg.inv(): 역 행렬(^-1)
# X.T: X의 전치 행렬
# np.dot(): 행렬 곱셈(dot product)
# X.dot(A): X와 A의 행렬 곱셈
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)

# plot을 통한 시각화
plt.scatter(X[:, 1], Y)
plt.plot(X[:, 1], X.dot(theta))
plt.show()
```

#### 예제 3

```python
x = [1, 2, 3, 4]
X = []
for i in x:
    X.append([1, i, i**2, i**3])
X = np.array(X)
Y = np.array([6, 2, 7, 4])
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)

# plot을 통한 시각화
plt.scatter(X[:, 1], Y)
x = np.linspace(1, 4, 100)
x_matrix = np.array([np.ones(100), x, x**2, x**3]).T
plt.plot(x, x_matrix.dot(theta))
plt.show()
```

#### 예제 4

```python
# 경사 하강법
X = np.array([[1, 1], [1, 2], [1, 3], [1, 4]])
Y = np.array([1.2, 1.8, 3.3, 4.1])

theta = np.random.randn(2)
learning_rate = 0.01

for i in range(100):
    grad = 2 * X.T.dot(X.dot(theta) - Y)
    theta = theta - learning_rate * grad
    
# plot을 통한 시각화
plt.scatter(X[:, 1], Y)
plt.plot(X[:, 1], X.dot(theta))
plt.show()
```

```python
cost = (Y - X.dot(theta)) ** 2
print(cost)
print(cost.sum() / len(X))
```
